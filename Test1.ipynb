```python
import keras
import tensorflow as tf
from keras.models import Sequential
from keras import optimizers
import pandas as pd
import numpy as np
import cv2
```

```python
new_model = tf.keras.models.load_model('EmotionRecognition101.model')
#new_model = tf.keras.models.load_model('EmotionRecognition88.model')
#new_model = tf.keras.models.load_model('EmotionRecognition95.model')
#new_model = tf.keras.models.load_model('EmotionRecognition100.model')
#new_model = tf.keras.models.load_model('EmotionRecognitionCK+D4.model')
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
new_model.summary()
```

```python
import numpy as np
import cv2

EMOTIONS = ["Happy","Sad","Angry","Fear","Surprise","Disgust","Neutral"];
#EMOTIONS = ["Neutral","Angry","Contempt","Disgust","Fear","Happy","Sad","Surprise"];

label= '';
cap = cv2.VideoCapture(0)

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi',fourcc, 20.0, (640,480))

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret==True:
        frame = cv2.flip(frame,1)
        img_data_list=[]
        # write the flipped frame
        out.write(frame)
        img=frame;
         #FaceCascade
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        #print(gray)
        #faces = face_cascade.detectMultiScale(gray, 1.3, 5)
        # Detect faces in the image
        faces = face_cascade.detectMultiScale(
            gray,
            scaleFactor=1.1,
            minNeighbors=5,
            minSize=(30, 30),
          
            )
        
        for (x,y,w,h) in faces:
            cv2.rectangle(gray,(x,y),(x+w,y+h),(255,0,0),2)
            cv2.putText(gray, label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.45, (0, 0,255), 2)
            img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
            roi_gray = gray[y:y+h, x:x+w]
            roi_color = img[y:y+h, x:x+w]
            cropped = img[y:y+h, x:x+w]    # To crop the image to get ROI.
            #cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
           
            
           
        
        #eyes = eye_cascade.detectMultiScale(roi_gray)
        #for (ex,ey,ew,eh) in eyes:
        
        img_to_yuv = cv2.cvtColor(img,cv2.COLOR_BGR2YUV)
        img_to_yuv[:,:,0] = cv2.equalizeHist(img_to_yuv[:,:,0])
        hist_equalization_result = cv2.cvtColor(img_to_yuv, cv2.COLOR_YUV2BGR)
        
        
        #Our keras model used a 4D tensor, (images x height x width x channel)
        #So changing dimension 128x128x3 into 1x128x128x3 
        img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        img_resize=cv2.resize(img,(200,200))
        img_data_list.append(img_resize)

        #img_data = np.array(array([array(Image.open(path2 + '\\'+'test'+'\\'+ imlist[0]))]))
        img_data = np.array(img_data_list);
        img_data = img_data.astype('float32')
        img_data /= 255
       
        #print(img_data[0])

        #plt.imshow(img_data[10])
        #plt.show()
        img_data= np.expand_dims(img_data, axis=4) 
        prediction =new_model.predict(img_data)
        emotion_probability = np.max(prediction)
        label = EMOTIONS[prediction.argmax()]
        print(label)
        #Calling the predict method on model to predict 'me' on the image
        #prediction = int(model.predict(img_array)[0][0])
        cv2.imshow('test_face', gray)
        #cv2.imshow('frame',frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break

# Release everything if job is finished
cap.release()
out.release()
cv2.destroyAllWindows()
```
